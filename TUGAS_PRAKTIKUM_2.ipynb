## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

## Membaca Dataset


```python
df = pd.read_csv(r"https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/1D_Porosity.csv") # data from Dr. Pyrcz's github repository
df.head()
```

## Informasi Dataset

```python
df.info()
```

## Statistik Deskriptif

```python
df.describe()
```

## Visualisasi Hubungan Variabel

```python
sns.scatterplot(x='Depth', y='Nporosity', data=df)
plt.show()
```

# Regresi Linier Sederhana
Model regresi:

Y = β0 + β1X + ε

## Menentukan Variabel

```python
X = df[['Depth']]
y = df['Nporosity']
```

## Membuat Model Regresi

```python
model = LinearRegression()
model.fit(X, y)
```

## Melihat Koefisien Model

```python
print('Intercept:', model.intercept_)
print('Koefisien:', model.coef_)
```

### model regresi
Y = −0.6919615 + 0.13935835 Depth

## Prediksi

```python
y_pred = model.predict(X)
print(y_pred)
```

##Membuat Visualisasi Model

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Depth', y='Nporosity', data=df, label='Data Asli')
plt.plot(X, y_pred, color='orange', label='Garis Regresi Orde 1')
plt.title('Scatter Plot Regresi Linear Sederhana (Depth vs Nporosity)')
plt.xlabel('Depth')
plt.ylabel('Nporosity')
plt.legend()
plt.grid(True)
plt.show()
```

## Scatter Plot Regresi Linear (Regresi Sederhana)

Penjelasan:

Scatter plot menunjukkan hubungan antara variabel Depth sebagai variabel independen dan Nporosity sebagai variabel dependen. Titik-titik pada grafik merupakan data asli, sedangkan garis regresi menunjukkan hasil prediksi dari model regresi linear.

Dari grafik terlihat bahwa garis regresi mencoba mengikuti pola distribusi data, namun karena hubungan data tidak sepenuhnya linear, beberapa titik masih cukup jauh dari garis regresi. Hal ini menyebabkan nilai MAE, MSE, dan RMSE tidak terlalu kecil. Nilai error tersebut menunjukkan bahwa masih terdapat selisih antara nilai aktual dan nilai prediksi model.

Semakin kecil nilai MAE, MSE, dan RMSE, maka model semakin baik dalam memprediksi data.

## Mean Absolute Error (MAE) , MEan Squared Error (MSE) dan Root Mean Square Error (RMSE)

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print('MAE:', mae)
print('MSE:',mse)
print('RMSE:',rmse)
```

# Regresi Polynomial

Regresi polynomial merupakan pengembangan dari regresi linier yang digunakan untuk memodelkan hubungan nonlinier antara variabel independen dan variabel dependen. Jika hubungan antara variabel tidak dapat dijelaskan dengan garis lurus, maka regresi polynomial dapat digunakan dengan menambahkan pangkat dari variabel independen ke dalam model.
